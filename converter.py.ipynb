<h1>... Making the Converter with a length of 20 using NLTK... </h1>

```python
import nltk
import string
```

```python
#Reading the file

f=open('big.txt')
input_str=f.read()
input_str=(input_str).lower()
len(input_str)
#input_str
```

```python
#Step-1 Removing the punctuations
# I am first replacing the punctuations with white space then removing multiple spaces

from nltk.tokenize import word_tokenize
import re

translator = re.compile('[%s]' % re.escape(string.punctuation))
input_str=translator.sub(' ', input_str)
input_str = re.sub(' +',' ', input_str).strip()
input_str[0:500]

```

```python
#Step-2 Tokenising and Removing stopwords

tokens=word_tokenize(input_str)

from nltk.tokenize import word_tokenize
from nltk.corpus import  stopwords
#nltk.download('stopwords')
stop_words=set(stopwords.words("english"))
#tokens=word_tokenize(input_str)
no_stopwords=[i for i in tokens if not i in stop_words]
no_stopwords
for ch in no_stopwords:
    if ch == 'this':
        print('hh')
```

```python
#step-3 Lemmatizing the words(Coverting plural to singular)

from nltk.stem import WordNetLemmatizer
#nltk.download("WordNetLemmatizer")
lemmatizer=WordNetLemmatizer()
no_lem=[]
for word in no_stopwords:
    no_lem.append(lemmatizer.lemmatize(word))
no_lem  
#len(no_lem)
```

```python
len(no_lem)
```

```python
#Removing the duplicacy because now duplicate words are of no use 

new_lem=set(no_lem)
len(new_lem)
```

```python
new_lem=(list(new_lem))
len(new_lem)
```

```python
#step-4 Forming Dictionary

word_dict={}
for i in range(len(new_lem)):
    word_dict[new_lem[i]]=i
print(word_dict) 
```

```python
#step-5 Making Converter

test_input=" fade project adventure are sherlock madinsane copyright arthur affairthis"
test_input=test_input.split()
test_input
```

```python
#step-6 Finding values of each word

final_vector=[]
# flag=0
for i in test_input:
    if word_dict.get(i) != None:
        final_vector.append(word_dict.get(i))
        print(i,word_dict.get(i))
        
    else:
        final_vector.append(0)
        print(i,0)

final_vector          
# len(final_vector)
```

```python
# So, the final returned vector of length 20 is:

# final_vector = final_vector+(final_vector + [0] * 4)[:len(final_vector)]
# final_vector
padding=[]
padding=padding+[0]*(20-len(final_vector))
final_vector+=padding
```
